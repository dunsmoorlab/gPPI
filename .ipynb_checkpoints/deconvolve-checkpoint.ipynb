```python
preproc = clean(bold,confounds=confounds.values,t_r=2,detrend=False,standardize='zscore')
preproc2= clean(bold,t_r=2,detrend=True,high_pass=1/128,standardize='zscore')
plt.figure(figsize=(20,5))
plt.plot(x,preproc);
plt.plot(x,preproc2)
plt.legend(['Preprocessed BOLD']);
```

```python
#set up parameters for deconvolution
N = preproc.shape[0]
TR = 2
hrf = np.array(_double_gamma_hrf(temporal_resolution=.5))
hrf = hrf / np.max(hrf)
plt.figure(figsize=(10,5))
plt.plot(hrf);
plt.legend(['double gamma HRF']);
```

```python
#make the cosine basis set
def dct_mat(N_,K_):
    n = np.array((range(0,N_))).T
    C_ = np.zeros((n.shape[0],K_))
    C_[:,0] = np.ones(n.shape[0])/np.sqrt(N_)
    for q in range(1,K_):
        C_[:,q] = np.sqrt(2/N_)* np.cos( np.pi*(2*n)* (q) /(2*N_)) #what is this???/?????
    return C_
```

```python
xb = dct_mat(N,N)
plt.figure(figsize=(10,10))
sns.heatmap(xb,cmap='mako_r',square=True);
plt.title('DCT Basis Set');
```

```python
#Convolve the basis set with the HRF
Hxb = np.zeros((N,N))
for i in range(N):
    Hx = convolve(xb[:,i],hrf)
    Hxb[:,i] = Hx[:preproc.shape[0]]
```

```python
plt.figure(figsize=(10,10))
sns.heatmap(Hxb,cmap='mako_r',square=True);
plt.title('Convolved Basis Set');
```

```python
reg = Ridge(alpha=1,solver='lsqr',fit_intercept=False,normalize=False,max_iter=1000)
reg.fit(Hxb,preproc)
neuronal = np.matmul(xb,reg.coef_[0,:])
plt.figure(figsize=(20,5))
plt.plot(x,neuronal)
plt.legend(['Neuronal signal'])
```

```python
recon = convolve(neuronal , hrf)[:bold.shape[0]]
plt.figure(figsize=(20,5))
plt.plot(x,preproc)
plt.plot(x,recon)
# plt.plot(x,neuronal)
plt.legend(['Original BOLD','Reconstructed BOLD']);
```
